```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mc

# read in csv as Pandas dataframe
df = pd.read_csv (r"C:\Users\katia\Downloads\en_climate_daily_ON_6105976_2021_P1D.csv")

# extract the relevant data
max_temps = df[['Days', 'Year', 'Max Temp (°C)']]
# change to Numpy array
MT = pd.DataFrame.to_numpy(max_temps)
# transpose
MT = MT.T

# set the colours for the temps
clrs = ['midnightblue' if (x <= -20) else 'indigo' if (-20 < x <= -16) else 'blueviolet' if (-16 < x <= -12) 
        else 'blue' if (-12 < x <= -8) else 'cornflowerblue' if (-8 < x <= -4) else 'c' if (-4 < x <= 0) 
        else 'forestgreen' if (0 < x <= 4) else 'mediumspringgreen' if (4 < x <= 8)
        else 'greenyellow' if (8 < x <= 12) else 'yellow' if (12 < x <= 16) 
        else 'lightsalmon' if (16 < x <= 20) else 'orange' if (20 < x <= 24)
        else 'red' if (24 < x <= 28) else 'firebrick' if (28 < x <= 32) else 'maroon' if (x > 32) else 'white' for x in MT[2,:] ]

# plot it
fig = plt.figure(figsize=(13,10))
ax = fig.subplots(1,1)
ax.bar(MT[0,:], MT[1,:], color = clrs) # plotting date (1-365) on x and year on y (to get uniform)
ax.set_xlabel("Month")
ax.set_xticks((1,31,59,90,120,151,181,212,243,273,304,334)) # space out ticks on x-axis for months
ax.set_xticklabels(("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul","Aug", "Sept", "Oct", "Nov", "Dec"))
ax.set_title("Temperature Blanket Test 2021")
plt.show()
```
